```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_student_data_final.csv")
print(df.head())
```

```python
print("Dataset Shape:", df.shape)
print("\nColumn Names:")
print(df.columns)

print("\nDataset Info:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df.duplicated().sum()
```

```python
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution of Students")
plt.show()
```

```python
plt.scatter(df['StudyHours'], df['FinalGrade'])
plt.title("Study Hours vs Final Grade")
plt.xlabel("Study Hours")
plt.ylabel("Final Grade")
plt.show()
```

```python
plt.scatter(df['Attendance'], df['FinalGrade'])
plt.title("Attendance vs Final Grade")
plt.xlabel("Attendance")
plt.ylabel("Final Grade")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])

sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.hist(df['FinalGrade'], bins=10)
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.show()
```

```python
top_students = df.sort_values(by='FinalGrade', ascending=False)
print(top_students[['Name','FinalGrade']].head(10))
```

```python
                       # INSIGHTS
    
#Students with higher study hours tend to get higher grades.

#Attendance has a positive relationship with final grades.

#Students with high parental support often perform better.

#Some students perform well even with low extracurricular activities.

#Online classes do not significantly affect final grades.
```
